## Popularity Based Recommendation System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
ratings.shape
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
print(num_rating_df)

num_rating_df.rename(columns={'Book-Rating':'num_ratings'} , inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df , on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(50)
```

```python
popular_df.merge(books , on='Book-Title').drop_duplicates('Book-Title')[['Book-Title' , 'avg_rating' ,'Book-Author', 'num_ratings' , 'Image-URL-M']].head(50)
```

## Collaborative Filtering Based Recommender system

```python
ratings_with_name
```

```python
x= ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
educated_user = x[x].index
educated_user
```

```python
filtered_rating =ratings_with_name[ratings_with_name['User-ID'].isin(educated_user)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
filtered_rating['Book-Title'].isin(famous_books)

final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]   ## Boolean Indexing
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title' , columns='User-ID' , values='Book-Rating')

## Only those users who have rated atleast 200 books
## Only those books which have been rated by atleast 50 users
```

```python
pt.shape
```

```python
pt.fillna(0 , inplace =True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
similarity_scores[0]
```

```python
similarity_scores[1]
```

```python
np.where(pt.index=='Zoya')[0][0]
np.where(pt.index=='Year of Wonders')[0][0]

```

```python
sorted(list(enumerate(similarity_scores[0])), key=lambda x:x[1], reverse=True)[1:6]
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items =sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:8]

    for i in similar_items:
        print(pt.index[i[0]])

    
```

```python
recommend('1984')
```
